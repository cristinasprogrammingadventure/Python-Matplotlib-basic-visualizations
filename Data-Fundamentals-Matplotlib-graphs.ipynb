```python
import pandas as pd
import matplotlib.pyplot as plt

peticions = pd.read_csv('Peticions ciutadanes_OpenDataBCN.csv')
peticions.head(10)
```

```python
#agrupamos por dos campos y nos quedamos solo con uno de los counts
peticions_mes= peticions.groupby(['TIPUS', 'MES_DATA_TANCAMENT'])['TIPUS'].count() 
peticions_mes
```

```python
#pivotamos la tabla para que el mes no sea un campo, sinó valores de columnas
peticions_pivot = peticions_mes.unstack('MES_DATA_TANCAMENT')
peticions_pivot
```

```python
colors = ['red', 'green']  # creamos una lista de dos colores
peticions_pivot.plot(kind="bar", figsize = (15,8), color=colors) #cremos el plot con el df anterior y con los colores
```

OTRA FORMA DE HACERLO


```python
#Filtramos por mes y creamos un df filtrado
pet_mes1=peticions[peticions['MES_DATA_TANCAMENT']==1.0]
pet_mes2=peticions[peticions['MES_DATA_TANCAMENT']==2.0]
```

```python
#agrupamos cada df
pet_mes1_group=pet_mes1.groupby(['TIPUS'])[['MES_DATA_TANCAMENT']].count()
pet_mes2_group=pet_mes2.groupby(['TIPUS'])[['MES_DATA_TANCAMENT']].count()
```

```python
#cambiamos nombre a la columna
pet_mes1_group.columns=['Gener']
pet_mes2_group.columns=['Febrer']
```

```python
pet_mes_concat=pd.concat([pet_mes1_group,pet_mes2_group],axis=1)
pet_mes_concat
```

```python
pet_mes_concat.plot(kind='barh',color=colors)
```
